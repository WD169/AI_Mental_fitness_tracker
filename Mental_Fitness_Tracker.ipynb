```python
#importing libraries

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
```

```python
#Extracting data:

data_1 = pd.read_csv("prevalence-by-mental-and-substance-use-disorder.csv")
data_2 = pd.read_csv("mental-and-substance-use-as-share-of-disease.csv")
```

```python
data_1.head()
```

```python
data_2.head()
```

```python
data = pd.merge(data_1,data_2)
data.head()
```

```python
#Data Cleaning:

data.isnull().sum()
```

```python
data.columns
```

```python
data.drop('Code',axis=1,inplace=True)
```

```python
data.set_axis(['Country','Year','Schizophrenia','Bipolar disorder','Eating disorders','Anxiety disorders','Drug use disorders','Depressive disorders','Alcohol use','Mental_Fitness'], axis='columns', inplace=True)
data.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
    if data[i].dtype == 'object':
        data[i]=l.fit_transform(data[i]) 
```

```python
data.shape
```

```python
#Spliting the dataset

X = data.drop('Mental_Fitness', axis=1)
y = data['Mental_Fitness']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.10, random_state=2)
```

```python
print("X_train",X_train.shape)
print("X_test",X_test.shape)
print("\ny_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
#Model training:

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lreg = LinearRegression()
lreg.fit(X_train, y_train)

y_train_pred = lreg.predict(X_train)
mse = mean_squared_error(y_train, y_train_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(y_train, y_train_pred)
```

```python
print("The Linear Regression model performance for training set")
print("---------------------------------------------------")
print("Mean Squared Error is ",format(mse))
print("Root Mean Squared Error is ",format(rmse))
print("R2 score is ",format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train, y_train)

y_train_pred = rf.predict(X_train)
mse = mean_squared_error(y_train, y_train_pred)
rmse = (np.sqrt(mse))
r2 = r2_score(y_train, y_train_pred)
```

```python
print("The Random Forest Regressor model performance for training set")
print("---------------------------------------------------")
print("Mean Squared Error is ",format(mse))
print("Root Mean Squared Error is ",format(rmse))
print("R2 score is ",format(r2))
```

```python
# Linear regression model evaluation for testing set

y_test_pred = lreg.predict(X_test) 
mse = mean_squared_error(y_test, y_test_pred)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_pred)))
r2 = r2_score(y_test, y_test_pred)

print("The Linear Regression model performance for testing set")
print("---------------------------------------------------")
print("Mean Squared Error is ",format(mse))
print("Root Mean Squared Error is ",format(rmse))
print("R2 score is ",format(r2))
```

```python
#Random Forest Regressor model evaluation for testing set

y_test_pred = rf.predict(X_test) 
mse = mean_squared_error(y_test, y_test_pred)
rmse = (np.sqrt(mean_squared_error(y_test, y_test_pred)))
r2 = r2_score(y_test, y_test_pred)

print("The Random Forest Regressor model performance for testing set")
print("---------------------------------------------------")
print("Mean Squared Error is ",format(mse))
print("Root Mean Squared Error is ",format(rmse))
print("R2 score is ",format(r2))
```
